# Run Details

## UDP
* All UDP code and excecutable files should be in the `/udp` folder.
### Client
* To run the UDP client use the following command: `./client <server_address> <save_file_name>`
* `server address` is the IP address of the server to query for time. It will be `0` for local machine.
* `save_file_name` is the file name to save timing data to. It will look something like `udp_output.txt`
* Sample run: `./client 0 udp_output.txt` or `./client 1.160.10.240 udp_output.txt`
### Server
* To run the UDP server use the following command: `./server`

## RPC
* All RPC code and excecutable files should be in the `/rpc` folder.
### Client
* To run the RPC client use the following command: `./clock_client <server_address> <save_file_name>`
* `server address` is the IP address of the server to query for time. It will be `localhost` for local machine.
* `save_file_name` is the file name to save timing data to. It will look something like `rpc_output.txt`
* Sample run: `./clock_client localhost udp_output.txt` or `./clock_client 1.160.10.240 udp_output.txt`
### Server
* To run the RPC server, use the following command: `./clock_server`

# Collect Data

```python
import numpy as np
import pandas as pd 
from datetime import datetime
import matplotlib.pyplot as plt

udp_same = pd.read_csv("udp/udp_same.txt",sep=',',header=0) 
rpc_same = pd.read_csv("rpc/rpc_same.txt",sep=',',header=0)
udp_diff_cu = pd.read_csv("udp/udp_diff_cu.txt",sep=',',header=0) 
rpc_diff_cu = pd.read_csv("rpc/rpc_diff_cu.txt",sep=',',header=0)
```

# Question 1

```python
def getLatencies(dataFrame):
    latencies = []
    for _,step in dataFrame.iterrows():
        clientSend = datetime.strptime(step['Client_Send'], '%H:%M:%S.%f')
        clientReceive = datetime.strptime(step['Client_Receive'], '%H:%M:%S.%f')
        latencie = clientReceive - clientSend
        latencies.append(latencie.total_seconds())
    return np.array(latencies)*10**6
```

```python
# udp same machine
udp_sameLatencies = getLatencies(udp_same)

# rpc same machine
rpc_sameLatencies = getLatencies(rpc_same)

# udp same machine
udp_diff_cuLatencies = getLatencies(udp_diff_cu)

# rpc same machine
rpc_diff_cuLatencies = getLatencies(rpc_diff_cu)
```

```python
fig = plt.figure()
fig.set_figheight(5)
fig.set_figwidth(15)

plt.subplot(1, 2, 1)
plt.plot(range(len(udp_sameLatencies)), udp_sameLatencies, alpha=0.5, color='r', label='UDP')
plt.plot(range(len(rpc_sameLatencies)), rpc_sameLatencies, alpha=0.8, color='b', label='RPC')
plt.xlabel('Data Points')
plt.ylabel('Latency (micro-seconds)')
plt.title('Latency per data point on the same machine')
plt.legend(loc="upper right")

plt.subplot(1, 2, 2)
plt.plot(range(len(udp_diff_cuLatencies)), udp_diff_cuLatencies, alpha=0.5, color='r', label='UDP')
plt.plot(range(len(rpc_diff_cuLatencies)), rpc_diff_cuLatencies, alpha=0.8, color='b', label='RPC')
plt.xlabel('Data Points')
plt.ylabel('Latency (micro-seconds)')
plt.title('Latency per data point on different machines on the same network')
plt.legend(loc="upper right")

plt.show()
```

### UDP on different machines on the same network

```python
def offset_delays(dataFrame):
    delays = []
    offsets = []
    for _,step in dataFrame.iterrows():
        ti_3 = datetime.strptime(step['Client_Send'], '%H:%M:%S.%f')
        ti_2 = datetime.strptime(step['Server_Receive'], '%H:%M:%S.%f')
        ti_1 = datetime.strptime(step['Server_Send'], '%H:%M:%S.%f')
        ti_0 = datetime.strptime(step['Client_Receive'], '%H:%M:%S.%f')
        delay = (ti_2 - ti_3) + (ti_0 - ti_1)
        offset = ((ti_2 - ti_3) - (ti_0 - ti_1))/2.0
        delays.append(delay.total_seconds())
        offsets.append(offset.total_seconds())
    return np.array(delays)*10**6, np.array(offsets)*10**6
```

```python
udp_same_delay, udp_same_offset = offset_delays(udp_same)
rpc_same_delay, rpc_same_offset = offset_delays(rpc_same)

udp_diff_cu_delay, udp_diff_cu_offset = offset_delays(udp_diff_cu)
rpc_diff_cu_delay, rpc_diff_cu_offset = offset_delays(rpc_diff_cu)
```

```python
fig = plt.figure()
fig.set_figheight(5)
fig.set_figwidth(15)

plt.subplot(1, 2, 1)
plt.plot(range(len(udp_same_delay)), udp_same_delay, alpha=0.5, color='r', label='UDP')
plt.plot(range(len(rpc_same_delay)), rpc_same_delay, alpha=0.8, color='b', label='RPC')
plt.xlabel('Data Points')
plt.ylabel('Delay (micro-seconds)')
plt.title('Delay per data point on the same machine')
plt.legend(loc="upper right")

plt.subplot(1, 2, 2)
plt.plot(range(len(udp_diff_cu_delay)), udp_diff_cu_delay, alpha=0.5, color='r', label='UDP')
plt.plot(range(len(rpc_diff_cu_delay)), rpc_diff_cu_delay, alpha=0.8, color='b', label='RPC')
plt.xlabel('Data Points')
plt.ylabel('Delay (micro-seconds)')
plt.title('Delay per data point on different machines on the same network')
plt.legend(loc="upper right")

plt.show()
```

```python
fig = plt.figure()
fig.set_figheight(5)
fig.set_figwidth(15)

plt.subplot(1, 2, 1)
plt.plot(range(len(udp_same_offset)), udp_same_offset, alpha=0.5, color='r', label='UDP')
plt.plot(range(len(rpc_same_offset)), rpc_same_offset, alpha=0.8, color='b', label='RPC')
plt.xlabel('Data Points')
plt.ylabel('Offset (micro-seconds)')
plt.title('Offset per data point on the same machine')
plt.legend(loc="upper right")

plt.subplot(1, 2, 2)
plt.plot(range(len(udp_diff_cu_offset)), udp_diff_cu_offset, alpha=0.5, color='r', label='UDP')
plt.plot(range(len(rpc_diff_cu_offset)), rpc_diff_cu_offset, alpha=0.8, color='b', label='RPC')
plt.xlabel('Data Points')
plt.ylabel('Offset (micro-seconds)')
plt.title('Offset per data point on different machines on the same network')
plt.legend(loc="upper right")

plt.show()
```

# Question 4

```python
def crisDiff(dataFrame):
    diffs = []
    for _,step in dataFrame.iterrows():
        ti_3 = datetime.strptime(step['Client_Send'], '%H:%M:%S.%f')
        ti_2 = datetime.strptime(step['Server_Receive'], '%H:%M:%S.%f')
        ti_1 = datetime.strptime(step['Server_Send'], '%H:%M:%S.%f')
        ti_0 = datetime.strptime(step['Client_Receive'], '%H:%M:%S.%f')
        t_new = ti_2 + (ti_0 - ti_3)/2.0
        diff = t_new-ti_0
        diffs.append(diff.total_seconds())
    return np.array(diffs)*10**6
```

```python
udp_same_diff = crisDiff(udp_same)
rpc_same_diff = crisDiff(rpc_same)

udp_diff_cu_diff = crisDiff(udp_diff_cu)
rpc_diff_cu_diff = crisDiff(rpc_diff_cu)
```

```python
fig = plt.figure()
fig.set_figheight(5)
fig.set_figwidth(15)

plt.subplot(1, 2, 1)
plt.plot(range(len(udp_same_diff)), udp_same_diff, alpha=0.5, color='r', label='UDP')
plt.plot(range(len(rpc_same_diff)), rpc_same_diff, alpha=0.8, color='b', label='RPC')
plt.xlabel('Data Points')
plt.ylabel('Difference (micro-seconds)')
plt.title('Difference per data point on the same machine')
plt.legend(loc="upper right")

plt.subplot(1, 2, 2)
plt.plot(range(len(udp_diff_cu_diff)), udp_diff_cu_diff, alpha=0.5, color='r', label='UDP')
plt.plot(range(len(rpc_diff_cu_diff)), rpc_diff_cu_diff, alpha=0.8, color='b', label='RPC')
plt.xlabel('Data Points')
plt.ylabel('Difference (micro-seconds)')
plt.title('Difference per data point on different machines on the same network')
plt.legend(loc="upper right")

plt.show()
```
